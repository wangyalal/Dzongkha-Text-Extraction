```python
list_with_ending = []
list_without_ending = []
with open("xi xii jungrab.txt", 'r', encoding='utf-8') as file:
    content = file.readlines()
    for line in content:
        if('།' in line):
            list_with_ending.append(line.strip())
        else:
            list_without_ending.append(line.strip())
```

```python
print(len(list_with_ending))
print(len(list_without_ending))
```

```python
for line in list_with_ending:
    print(line)
```

```python
for line in list_without_ending:
    print(line)
```
